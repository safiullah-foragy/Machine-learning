# Cramers rule

```python
import numpy as np

A = np.array([[1,2,3],
              [2,3,1],
              [3,1,2]], dtype=float)
b = np.array([1,2,3], dtype=float)

solution = np.linalg.solve(A, b)
print("Q1 Solution (weights A,B,C):", solution)
```

# Gauss jordan

```python
A = np.array([[2,-2,3],
              [1,2,-1],
              [3,-1,2]], dtype=float)
b = np.array([2,3,1], dtype=float)

solution = np.linalg.solve(A, b)
print("Q5 Solution (x,y,z):", solution)
```
